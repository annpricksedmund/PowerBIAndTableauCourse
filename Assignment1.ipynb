− Power BI Desktop

Power BI Desktop is the free version of Power BI that you can install on your local computer as a program, and acts as a companion desktop application to the full version of Power BI. <br>
Despite its free nature, you may be surprised at what Microsoft has included with Power BI Desktop:<br>
•	You can connect and import data from over 70 cloud-based and on-premises sources<br>
•	The same rich visualisations and filters from Power BI Pro<br>
•	Auto-detect that finds and creates data relationships between tables and formats<br>
•	Export your reports to CSV, Microsoft Excel, Microsoft PowerPoint and PDF<br>
•	Python support<br>
•	Save, upload and publish your reports to the Web and the full Power BI service<br>
•	Storage limit of 10 GB per user<br>
Naturally, there are some feature limitations with the free Power BI Desktop that give further incentive to upgrade.<br>
•	Can’t share created reports with non-Power BI Pro users<br>
•	No App Workspaces<br>
•	No API embedding<br>
•	No email subscriptions<br>
•	No peer-to-peer-sharing<br>
•	No support to analyse in Excel within Power BI Desktop

− Power BI Pro

Power BI Pro is the full version of Power BI, complete with the ability to use Power BI for both building dashboards and reports and unlimited viewing, sharing and consumption of your created reports (and reports shared by others) - the latter not possible with Power BI Desktop.<br>

A breakdown of Power BI Pro’s differences against Power BI Desktop:<br>
•	Ability to embed Power BI visuals into apps (PowerApps, SharePoint, Teams, etc)<br>
•	Native integration with other Microsoft solutions (Azure Data Services)<br>
•	Share datasets, dashboards and reports with other Power BI Pro users<br>
•	Can create App Workspaces and peer-to-peer sharing<br>
<br>
<br>
Price:- $9.99 (Monthly price per user)

−  Power BI Premium

Power BI Premium is the most expensive tier of Power BI currently available and very distinct from the other two versions available on the market.<br>
On top of the features and functionality standard to all versions of the service, users of Power BI Premium get:<br>
•	Increased data capacity limits and maximum performance<br>
•	Access to one API surface<br>
•	Ability to embed Power BI visuals into apps (PowerApps, SharePoint, Teams, etc)<br>
•	Larger storage sizes for extended deployments<br>
•	Geo distribution, higher refresh rates, isolation, pin to memory, read-only replicas<br>
•	Power BI Report Server<br>
<br>
<br>
Price: - $4995 (Monthly price per dedicated cloud compute and storage resource with annual subscription)
